```python
from pandas import read_csv
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import gradient_descent_v2
from sklearn.preprocessing import LabelEncoder
from keras.callbacks import LearningRateScheduler
```

# Planificación de la tasa de aprendizaje con disminución basada en caídas

```python
dataframe = read_csv("cd.csv", header=0)
datos = dataframe.values
```

```python
# planificacion de tasa de aprendizaje
def paso_decaimiento(epoca):
  tasa_apre_ini = 0.1
  disminucion = 0.5 # disminución al 50%
  epocas_dis = 10.0
  tasa_apre = tasa_apre_ini * math.pow(disminucion,
  math.floor((1+epoca)/epocas_dis))
  return tasa_apre
```

```python
X = datos[:,0:10].astype(float)
Y = datos[:,10]
```

```python
codificador = LabelEncoder()
codificador.fit(Y)
Y = codificador.transform(Y)
```

```python
# creacion del modelo
modelo = Sequential()
modelo.add(Dense(35, input_dim=10, activation='relu'))
modelo.add(Dense(1, activation='sigmoid'))
```

```python
#compilar modelo
sgd = gradient_descent_v2.SGD(learning_rate=0.0,momentum=0.9)
modelo.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
# callback de planificacion de aprendizaje
tasa_aprendizaje = LearningRateScheduler(paso_decaimiento)
lista_callbacks = [tasa_aprendizaje]
```

```python
# entrenar modelo
modelo.fit(X, Y, validation_split=0.33, epochs=50, batch_size=28, callbacks=lista_callbacks, verbose=2)
```
