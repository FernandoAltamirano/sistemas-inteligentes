```python
from pandas import read_csv
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import gradient_descent_v2
from sklearn.preprocessing import LabelEncoder
```

# Programación de la tasa de aprendizaje basada en el tiempo

```python
dataframe = read_csv("cd.csv", header=0)
datos = dataframe.values
```

```python
dataframe.head()
```

```python
dataframe.pop("Unnamed: 0")
```

```python
dataframe.head()
```

```python
X = datos[:,0:10].astype(float)
Y = datos[:,10]
```

```python
codificador = LabelEncoder()
codificador.fit(Y)
Y = codificador.transform(Y)
```

```python
# creacion del modelo
modelo = Sequential()
modelo.add(Dense(35, input_dim=10, activation='relu'))
modelo.add(Dense(1, activation='sigmoid'))
```

```python
#compilar modelo
epocas = 100
tasa_aprendizaje = 0.1
tasa_decaimiento = tasa_aprendizaje / epocas
momento = 0.8
sgd = gradient_descent_v2.SGD(learning_rate=tasa_aprendizaje,
momentum=momento, decay=tasa_decaimiento,
nesterov=False)
modelo.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
# entrenar modelo
modelo.fit(X, Y, validation_split=0.33, epochs=epocas, batch_size=28, verbose=2)
```
